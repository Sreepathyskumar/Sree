```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Q9a = pd.read_csv("C:/Users/subhash k/OneDrive/Desktop/File for assignment/1/Q9_a.csv")
Q9a
```

```python
skew_speed = Q9a['speed'].skew()
kurtosis_speed = Q9a['speed'].kurtosis()

skew_distance = Q9a['dist'].skew()
kurtosis_distance = Q9a['dist'].kurtosis()

print("Skewness for speed:", skew_speed)
print("Kurtosis for speed:", kurtosis_speed)
print("Skewness for dist:", skew_distance)
print("Kurtosis for dist:", kurtosis_distance)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(Q9a['speed'], kde=True)
plt.title('speed Distribution')

plt.subplot(1, 2, 2)
sns.histplot(Q9a['dist'], kde=True)
plt.title('dist Distribution')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean, cov = [0, 0], [(1, .6), (.6, 1)]
x, y = np.random.multivariate_normal(mean, cov, 100).T
y += x + 1

f, ax = plt.subplots(figsize=(6, 6))

ax.scatter(x, y, c=".3")
ax.set(xlim=(-3, 3), ylim=(-3, 3))
diag_line, = ax.plot(ax.get_xlim(), ax.get_ylim(), ls="--", c=".3")

def on_change(axes):
    x_lims = ax.get_xlim()
    y_lims = ax.get_ylim()
    diag_line.set_data(x_lims, y_lims)
ax.callbacks.connect('xlim_changed', on_change)
ax.callbacks.connect('ylim_changed', on_change)

plt.show()
```

```python
plt.hist(Q21a["MPG"], bins = 20, edgecolor=  'black')
plt.show()
```
